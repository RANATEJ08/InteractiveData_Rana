## Group_1 Titanic seaborn challenge 

```python
import pandas as pd 
import seaborn as sns
```

```python
titanic = pd.read_csv('titanic_clean.csv')
```

```python
titanic.head()
```

```python

sns.relplot(
    data= titanic,
    x= "age", y= "fare")
```

```python
sns.relplot(
    data= titanic,
    x= "age", y= 'fare',
    alpha = 0.05)
```

```python
sns.relplot(
    data= titanic,
    x= "age", y= 'fare',
    alpha = 0.8,hue= 'sex')
```

```python
sns.relplot(
    data= titanic,
    x= "age", y= 'fare',
    alpha = 0.8,hue= 'sex',
    col = 'survived')
```

```python
sns.relplot(
    data=titanic,
    x="age", y="fare", col="survived",
    hue="sex", size="pclass",
    palette={"male": "blue", "female": "red"},
)    
```

```python
titanic['sex'] = titanic['sex'].map({"male": "male", "female": "female"})
                                    
sns.relplot(
    data=titanic,
    x="age", y="fare", col="survived",
    hue="sex", 
    palette={"male": "blue", "female": "red"},
 )
```

```python
sns.relplot(
    data=titanic,
    x="age", y="fare", col="survived",
    hue="sex", 
    palette={"male": "blue", "female": "red"},
    size="pclass"
)    
```

```python
titanic['survive_ind'] = titanic['survived'].map({0: "didn't survive", 1: "survived"})


sns.relplot(
    data=titanic,
    x="fare", y="age", col="sex",
    hue="survive_ind", size="pclass",
    palette={"didn't survive": "red", "survived": "black"},
    alpha = 0.4
)
```

## Group_2  student mental health




```python
import pandas as pd 
from matplotlib import pyplot as plt 
import seaborn as sns 
```

```python
smh = pd.read_csv("Student Mental health.csv")
smh.head()
```

```python
smh.groupby("Choose your gender")[['Do you have Depression?','Do you have Anxiety?','Do you have Panic attack?']].count()
```

```python
column_to_plot = smh.columns[-4:-1]

category_order = ['Yes','No']
```

```python
for column in column_to_plot:
    counts = smh[column].value_counts()
    counts = counts.reindex(category_order, fill_value=0)
    print(counts)
    print()
```

```python
sns.countplot(smh, x= 'Do you have Depression?', stat='percent',hue=  'Did you seek any specialist for a treatment?')
```

```python
sns.countplot(smh, x= 'Do you have Anxiety?', stat='percent',hue=  'Did you seek any specialist for a treatment?')
```

```python
sns.countplot(smh, x= 'Do you have Panic attack?', stat='percent',hue=  'Did you seek any specialist for a treatment?')
```

```python
sns.countplot(smh, x= 'Do you have Depression?', stat='percent', hue =  'Choose your gender', palette = 'PuBuGn')
```
